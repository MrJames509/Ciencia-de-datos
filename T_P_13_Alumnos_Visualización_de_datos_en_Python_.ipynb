```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# 1. Cargamos el archivo usando read_excel porque termina en .xlsx
df = pd.read_excel('/content/ventas_sucio.xlsx')

# 2. Configuramos el tamaño de la figura (sin la palabra csv)
plt.figure(figsize=(10, 6))
sns.set_theme(style="whitegrid")

# 3. Agrupamos y ordenamos los datos
recaudacion_rubro = (
    df.groupby('Rubro')['Importe']
    .sum()
    .reset_index()
    .sort_values(by='Importe', ascending=False)
)

# 4. Creamos el gráfico de barras
sns.barplot(data=recaudacion_rubro, x='Importe', y='Rubro', palette='Blues_r')

plt.title(
    'Recaudación Total por Rubro', fontsize=14, fontweight='bold', pad=15
)
plt.xlabel('Recaudación Total ($)', fontsize=12)
plt.ylabel('Rubro', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5))

# Creamos el histograma
sns.histplot(data=df, x='Importe', bins=30, kde=True, color='teal')

plt.title('Distribución de Rangos de Precios de Ventas', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Importe de la Venta ($)', fontsize=12)
plt.ylabel('Frecuencia (Cantidad de Ventas)', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))

# Creamos el boxplot
sns.boxplot(data=df, x='Sede', y='Satisfaccion_Cliente', palette='Set2')

plt.title('Evaluación de Satisfacción del Cliente por Sede', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Sede', fontsize=12)
plt.ylabel('Nivel de Satisfacción', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))

# Creamos el gráfico de dispersión con un toque de transparencia (alpha) por si se solapan los puntos
sns.scatterplot(data=df, x='Importe', y='Comision_x_venta', alpha=0.7, color='coral')

plt.title('Impacto del Importe de Venta en la Comisión del Vendedor', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Importe de la Venta ($)', fontsize=12)
plt.ylabel('Comisión por Venta ($)', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Configuración estética general
sns.set_theme(style="whitegrid")

# Cargamos el archivo .xlsx de la carpeta de Colab
df = pd.read_excel('/content/ventas_sucio.xlsx')
print("¡Datos cargados con éxito! Dimensiones del dataset:", df.shape)
```

```python
plt.figure(figsize=(10, 6))

# Agrupamos por Rubro, sumamos el Importe y ordenamos de mayor a menor
recaudacion_rubro = (
    df.groupby('Rubro')['Importe']
    .sum()
    .reset_index()
    .sort_values(by='Importe', ascending=False)
)

# Creamos el gráfico de barras horizontales
sns.barplot(data=recaudacion_rubro, x='Importe', y='Rubro', palette='Blues_r')

plt.title(
    'Recaudación Total por Rubro', fontsize=14, fontweight='bold', pad=15
)
plt.xlabel('Recaudación Total ($)', fontsize=12)
plt.ylabel('Rubro', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5))

# Creamos el histograma con curva de densidad (kde)
sns.histplot(data=df, x='Importe', bins=30, kde=True, color='teal')

plt.title(
    'Distribución de Rangos de Precios de Ventas',
    fontsize=14,
    fontweight='bold',
    pad=15,
)
plt.xlabel('Importe de la Venta ($)', fontsize=12)
plt.ylabel('Frecuencia (Cantidad de Ventas)', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))

# Creamos el boxplot
sns.boxplot(data=df, x='Sede', y='Satisfaccion_Cliente', palette='Set2')

plt.title(
    'Evaluación de Satisfacción del Cliente por Sede',
    fontsize=14,
    fontweight='bold',
    pad=15,
)
plt.xlabel('Sede', fontsize=12)
plt.ylabel('Nivel de Satisfacción', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))

# Creamos el boxplot
sns.boxplot(data=df, x='Sede', y='Satisfaccion_Cliente', palette='Set2')

plt.title(
    'Evaluación de Satisfacción del Cliente por Sede',
    fontsize=14,
    fontweight='bold',
    pad=15,
)
plt.xlabel('Sede', fontsize=12)
plt.ylabel('Nivel de Satisfacción', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))

# Creamos el scatter plot con transparencia para evitar solapamiento
sns.scatterplot(
    data=df, x='Importe', y='Comision_x_venta', alpha=0.7, color='coral'
)

plt.title(
    'Impacto del Importe de Venta en la Comisión del Vendedor',
    fontsize=14,
    fontweight='bold',
    pad=15,
)
plt.xlabel('Importe de la Venta ($)', fontsize=12)
plt.ylabel('Comisión por Venta ($)', fontsize=12)
plt.tight_layout()
plt.show()
```
