##Ejercicio 1 A: El Desafío del Inventario Tecnológico

Contexto: Son Data Scientists en una empresa de tecnología y reciben datos de sensores de 5 servidores distintos. Los datos están crudos en un array de NumPy.

Etapa 1: Construcción "Artesanal" (Manual)
Consigna: Deben crear un DataFrame llamado df_manual utilizando el método de segmentación de un array maestro (Upcasting).

Creen un array de NumPy que contenga en la primera fila los nombres de las columnas y en la primera columna los nombres de las filas.

El array debe tener 5 filas de datos y 3 columnas de datos (sin contar etiquetas).

Restricción: Deben usar data2[1:,1:] para los datos y las posiciones específicas para index y columns.


Etapa 2: Construcción Profesional (Directa)
Consigna: Ahora, utilicen el método más eficiente de Pandas para crear df_profesional usando el array de datos puros y los parámetros data, index y columns en una sola línea.

```python
import numpy as np
import pandas as pd

# 1. Este es el array que deben procesar
datos_it = np.array([
    [95.5, 40, 1.2],
    [88.0, 35, 1.5],
    [np.nan, 42, 2.0],
    [75.2, 38, 1.1],
    [92.1, 45, 1.8]
])

# Etiquetas que DEBEN usar:
columnas = ['CPU_%', 'RAM_%', 'Latencia']
filas = ['Server_A', 'Server_B', 'Server_C', 'Server_D', 'Server_E']
```

```python
# 1. Creamos la matriz vacía de strings (6 filas x 4 columnas en total)
data2 = np.empty((6, 4), dtype='<U32')

# 2. Asignamos la esquina superior izquierda (vacía o con un placeholder)
data2[0, 0] = ''

# 3. Colocamos los nombres de las columnas en la primera fila
data2[0, 1:] = columnas

# 4. Colocamos los nombres de las filas en la primera columna
data2[1:, 0] = filas

# 5. Insertamos los datos puros en el bloque restante
data2[1:, 1:] = datos_it

# 6. Construimos df_manual segmentando el array maestro según la restricción
df_manual = pd.DataFrame(
    data=data2[1:, 1:],    # Restricción obligatoria: los datos puros
    index=data2[1:, 0],    # Primera columna (nombres de servidores)
    columns=data2[0, 1:]   # Primera fila (nombres de métricas)
)

print("--- DataFrame Manual (Etapa 1) ---")
print(df_manual)
```

```python
# Construcción directa y limpia en una sola línea
df_profesional = pd.DataFrame(data=datos_it, index=filas, columns=columnas)

print("\n--- DataFrame Profesional (Etapa 2) ---")
print(df_profesional)
```
