```python
import pandas as pd

df = pd.read_excel("/ventas_sucio.xlsx")

print(df.head())      # primeras filas
print(df.info())      # tipos de datos y nulos
df.shape      # cantidad de filas y columnas
```

```python
cantidad_duplicados = df.duplicated().sum()
print(f"Duplicados encontrados: {cantidad_duplicados}")

df = df.drop_duplicates()
print(f"Nuevo tamaño: {df.shape}")
```

```python
print(df.isnull().sum())
```

```python
columnas_numericas = df.select_dtypes(include="number").columns

for col in columnas_numericas:
    if df[col].isnull().sum() > 0:
        promedio = df[col].mean()
        df[col] = df[col].fillna(promedio)
        print(f"'{col}' rellenada con el promedio: {promedio:.2f}")
```

```python
columnas_categoricas = df.select_dtypes(exclude="number").columns

for col in columnas_categoricas:
    if df[col].isnull().sum() > 0:
        moda = df[col].mode()[0]
        df[col] = df[col].fillna(moda)
        print(f"'{col}' rellenada con la moda: '{moda}'")
```

```python
print(df.isnull().sum())          # debe dar todo en 0
df.to_csv("ventas_limpio.csv", index=False)
```

```python
# Respaldo de la columna original antes de tocarla (columna de seguridad)
df['Fecha_Venta_original'] = df['Fecha_Venta']

# Convertimos el texto a fecha real
# errors='coerce' convierte lo que no se puede interpretar en NaT (fecha nula)
df['Fecha_Venta'] = pd.to_datetime(df['Fecha_Venta'], dayfirst=True, errors='coerce')

print(df[['Fecha_Venta_original', 'Fecha_Venta']].head(10))
print(df['Fecha_Venta'].isnull().sum(), "fechas no se pudieron convertir")
```

```python
# Nos aseguramos que Importe sea numérico (float)
df['Importe'] = pd.to_numeric(df['Importe'], errors='coerce').astype(float)

# Configuramos el formato de visualización de Pandas: signo $, separador de miles y 2 decimales
pd.options.display.float_format = '${:,.2f}'.format

print(df['Importe'].head())
```

```python
print(df['Importe'].describe())
```

```python
# Caso A: valores negativos -> error imposible, se eliminan
negativos = df[df['Importe'] < 0]
print(f"Ventas con importe negativo: {len(negativos)}")

df = df[df['Importe'] >= 0]

# Caso B: valores extremadamente altos -> los identificamos para debatir,
# no los borramos automáticamente
umbral_alto = df['Importe'].mean() + 3 * df['Importe'].std()
outliers_altos = df[df['Importe'] > umbral_alto]

print(f"Ventas inusualmente altas (posibles outliers): {len(outliers_altos)}")
print(outliers_altos[['Importe']])

# Ejemplo de decisión: si el equipo decide que son errores, se filtran así:
# df = df[df['Importe'] <= umbral_alto]
```

```python
# Vemos las variantes que hay antes de corregir
print(df['Rubro'].unique())

# Unificamos todo a formato "Título" (primera letra de cada palabra en mayúscula)
df['Rubro'] = df['Rubro'].str.strip().str.title()

print(df['Rubro'].unique())
```

```python
# Vemos qué valores únicos tiene la columna antes de mapear
print(df['Prioridad_Pedido'].unique())

# Creamos el diccionario de equivalencias
mapa_prioridad = {
    'Baja': 1,
    'Media': 2,
    'Alta': 3
}

# Aplicamos el mapeo para convertir texto en números
df['Prioridad_Pedido_num'] = df['Prioridad_Pedido'].map(mapa_prioridad)

print(df[['Prioridad_Pedido', 'Prioridad_Pedido_num']].head(10))
```

```python
reporte_por_rubro = df.groupby('Rubro').agg(
    recaudacion_total=('Importe', 'sum'),
    satisfaccion_moda=('Satisfaccion_Cliente', lambda x: x.mode()[0])
)

print(reporte_por_rubro)
```

```python
reporte_por_rubro_sede = df.groupby(['Rubro', 'Sede']).agg(
    recaudacion_total=('Importe', 'sum'),
    cantidad_ventas=('Importe', 'count')
)

print(reporte_por_rubro_sede)
```
