```python
import pandas as pd
import numpy as np
```

```python
peliculas = pd.DataFrame(
            {'Año':[2014, 2014, 2013, 2013, 2001],
             'Valoración':[6, None, 8.75, None, 8.9],
             'Presupuesto':[160, 250, 100, None, 93],
             'Director':['Gareth Edwards', 'Peter Jackson', 'Martin Scorsese', 'Alfonso Cuarón', 'Peter Jackson'],
             'Título':['Godzilla', 'El Hobbit III', 'El lobo de Wall Street', 'Gravity', 'Lord of the Rings']}
)
peliculas
```

```python
peliculas['Recaudacion'] = [400, 850, 300, 1100, 950]
```

```python
# Generamos y guardamos la tabla dinámica según los requisitos
tabla_taquilla = pd.pivot_table(
    peliculas,
    values='Recaudacion',
    index='Director',
    columns='Año',
    aggfunc='sum',
    fill_value="-"
)

# Mostramos el resultado en pantalla
tabla_taquilla
```

```python
# Generamos y guardamos la tabla dinámica según los requisitos
tabla_taquilla = pd.pivot_table(
    peliculas,
    values='Recaudacion',
    index='Director',
    columns='Año',
    aggfunc='sum',
    fill_value="-"
)

# Mostramos el resultado en pantalla
tabla_taquilla
```

```python
# 1. Ordenar columnas cronológicamente (Eje de columnas = 1)
tabla_taquilla.sort_index(axis=1, inplace=True)

# 2. Ordenar filas de la Z a la A (Eje de filas = 0)
tabla_ordenada_z_a = tabla_taquilla.sort_index(axis=0, ascending=False)
tabla_ordenada_z_a
```

```python
# Volvemos a generar la tabla pivot pero con ceros para poder ordenar numéricamente
tabla_ceros = pd.pivot_table(
    peliculas,
    values='Recaudacion',
    index='Director',
    columns='Año',
    aggfunc='sum',
    fill_value=0
)

# Ordenamos las filas según la columna del año 2014 de mayor a menor
tabla_por_2014 = tabla_ceros.sort_values(by=2014, ascending=False)
tabla_por_2014
```

```python
complejo_A = pd.Series([2, 4, 0], index=['Peli_1', 'Peli_2', 'Peli_3'])
complejo_B = pd.Series([1, 2], index=['Peli_1', 'Peli_2'])

# Realizamos la división con fill_value=0
resultado_div = complejo_A.div(complejo_B, fill_value=0)
resultado_div
```
