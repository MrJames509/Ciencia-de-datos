##DESAFÍO: "Mini-Inventario de Aula"

Consigna:

Tenemos una lista de materiales y una tupla con sus cantidades. Vamos a unirlos y realizar una serie de operaciones para practicar la gestión de diccionarios.


1-Unir: Creá el diccionario materiales uniendo la lista nombres y la tupla cantidades con la función zip.


2-Consultar: Imprimí cuántas Hojas hay en el inventario utilizando dos formas: mediante corchetes [] y con el método .get().


3-Modificar: Cambiá la cantidad de Lapiceras a 25 directamente (usando el acceso por corchetes []). Imprimir para verificar.


4-Agregar: Utilizá el método .update() para incorporar el siguiente par al diccionario: {"Reglas": 5}.


5-Borrar: Eliminá el material Gomas del inventario usando la instrucción del.Imprimir para ver .


6-Ver: Mostrá el resultado final convirtiendo el diccionario en una lista de ítems con list(materiales.items())

```python
# DATOS
nombres = ["Lapiceras", "Hojas", "Gomas"]
cantidades = (10, 500, 8)

# RESOLVER ACÁ

# 1. Crear el diccionario
# materiales = ...
materiales = dict(zip(nombres, cantidades))

# 2. Ver cuántas Hojas hay
# print(...)
print("Cantidad de Hojas (con []):", materiales["Hojas"])
print("Cantidad de Hojas (con .get()):", materiales.get("Hojas"))

# 3. Cambiar Lapiceras a 25
materiales["Lapiceras"] = 25
print("Verificación de Lapiceras:", materiales["Lapiceras"])

# 4. Agregar Reglas con .update()
materiales.update({"Reglas": 5})

# 5. Borrar Gomas con del
del materiales["Gomas"]
print("Diccionario actual después de borrar Gomas:", materiales)

# 6. Mostrar resultado final como lista de tuplas
print("Resultado final:", list(materiales.items()))
```

##DESAFÍO:  Control de Gremios en Obra

Consigna:

Tenemos dos listas de especialistas que trabajaron en diferentes etapas de la obra. Vamos a usar los Sets para organizar la información sin que nada se repita y hacer un control de personal.

Pasos a seguir:


1-Limpiar y Convertir: Creá dos conjuntos llamados obra_A y obra_B convirtiendo las listas de arriba con set().
(Fijate que "Albañil" estaba repetido, ¡el Set lo va a limpiar solo!).



2-Unir (Total de Gremios): Usá el método .union() para crear un tercer conjunto llamado plantel_total que junte a todos los especialistas de ambas etapas sin repetir ninguno.


3-Consultar (Intersección): ¿Qué gremios trabajaron en ambas etapas? Usá .intersection() para descubrirlo e imprimí el resultado.


4-Diferencia (Nuevos): Queremos saber qué gremios se sumaron recién en la etapa_2 (es decir, los que están en la B pero NO estaban en la A).
Usá el signo menos (-) o .difference().


5-Modificar (Agregar): Resulta que hoy se sumó un "Gasista". Usá .add() para agregarlo al plantel_total.


6-Borrar (Retiro): El "Pintor" ya terminó su trabajo. Usá .remove() para sacarlo del plantel_total.


7-Ver Final: Imprimí el conjunto plantel_total para ver cómo quedó la lista definitiva de la obra.

```python
# DATOS (Copiá y pegá estas listas para empezar)
etapa_1 = ["Plomero", "Albañil", "Electricista", "Albañil"]
etapa_2 = ["Pintor", "Plomero", "Zinguero"]

# RESOLVER ACÁ

# 1. Limpiar y Convertir a Sets (Borrando repetidos automáticamente)
obra_A = set(etapa_1)
obra_B = set(etapa_2)

# 2. Unir ambos grupos sin repetir nada (El plantel total)
plantel_total = obra_A.union(obra_B)

# 3. Ver quiénes trabajaron en AMBAS etapas (Intersección)
print("Trabajaron en ambas etapas:", obra_A.intersection(obra_B))

# 4. Ver quiénes son "Nuevos" en la etapa_2 (Diferencia: B menos A)
nuevos = obra_B.difference(obra_A)
print(nuevos)

# 5. Agregar un "Gasista" al plantel_total
plantel_total.add("Gasista")

# 6. Borrar al "Pintor" porque ya terminó
plantel_total.remove("Pintor")

# 7. Mostrar el resultado final
print(f"Lista definitiva de la obra: {plantel_total}")
```

##Ejercicio Integrador: Gestión de Asistencia y Calificaciones

Contexto: Tenemos una lista de nombres de alumnos y una tupla con sus notas finales. Debemos organizar esta información en un diccionario, realizar actualizaciones de último momento y limpiar los datos.


Consigna:


Creación: Uní la lista alumnos y la tupla notas en un diccionario llamado registro_clase usando zip.


Consulta Segura: Usá el método .get() para buscar la nota de "Marcos". (Si no existe, el programa no debe dar error).


Slicing: Creá una nueva lista llamada primeros_dos que contenga solo los nombres de los dos primeros alumnos de la lista original.


Actualización: Un alumno nuevo se sumó tarde. Usá .update() para agregar a {"Esteban": 9} al registro_clase.


Modificación Directa: "Ana" mejoró su nota. Cambiá su valor a 10 usando el acceso directo por corchetes [].


Extracción con .pop(): El alumno "Luis" se dio de baja. Usá .pop() para borrarlo del diccionario y guardá su nota en una variable llamada nota_retirado.


Verificación: Comprobá si "Luis" todavía está en el diccionario usando not in (imprimí el resultado booleano).


Visualización Final: Imprimí el diccionario final convertido en una lista de tuplas usando .items().

```python
# --- DATOS INICIALES (Proporcionados por la Profe) ---
alumnos = ["Ana", "Luis", "Marta", "Juan"]
notas = (8, 7, 9, 6)

# --- ESPACIO PARA EL ALUMNO ---

# 1. Crear el diccionario registro_clase
registro_clase = dict(zip(alumnos, notas))

# 2. Buscar nota de "Marcos" con .get()
print("Nota de Marcos:", registro_clase.get("Marcos"))

# 3. Slicing de los primeros 2 nombres de la lista 'alumnos'
primeros_dos = alumnos[0:2]

# 4. Agregar a Esteban con .update()
registro_clase.update({"Esteban": 9})

# 5. Cambiar nota de Ana a 10 usando []
registro_clase["Ana"] = 10

# 6. Sacar a Luis con .pop() y guardar su nota
nota_retired = registro_clase.pop("Luis")

# 7. Verificar con 'not in' si Luis ya no está
print("¿Luis ya no está en el registro?:", "Luis" not in registro_clase)

# 8. Mostrar el registro final como lista de tuplas
print("Registro final:", list(registro_clase.items()))
```
