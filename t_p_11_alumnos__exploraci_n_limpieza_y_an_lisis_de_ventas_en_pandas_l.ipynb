🧠 Antes de avanzar: ¿Qué hace realmente .dropna()?

Es súper común confundirse y pensar que .dropna() borra columnas enteras o datos válidos. ¡Para nada! El comando .dropna() siempre busca eliminar únicamente las FILAS que contienen valores vacíos (NaN o None).


La clave está en si usamos o no el argumento subset:


1. Si NO usamos subset ➡️ mufi.dropna() Le decimos a Python: "Mirá TODA la tabla. Si una fila tiene un vacío en el Importe, o en el Vendedor, o en la Satisfacción... donde sea... ¡me borrás la fila entera!".

⚠️ Peligro: Podés terminar borrando una fila con una venta real de $2.000.000 solo porque el cliente se olvidó de poner una puntuación en la columna de satisfacción.

2. Si SÍ usamos subset ➡️ mufi.dropna(subset=['Importe']) Le ponemos filtro a Pandas y le decimos: "Mirá ÚNICAMENTE la columna Importe. Si encontrás un vacío ahí, borrá la fila entera. Si hay un número real (como los mayores a $350.000), la fila SE QUEDA intacta. Y si hay vacíos en otras columnas... ¡ignoralos!".


📌 En resumen: Los números reales mayores a $350.000 no se borran jamás. El subset solo le avisa a Python en qué columna tiene que mirar para buscar los baches antes de decidir si elimina la fila o no.

Si no usás .copy(), la variable mufi no es una tabla nueva e independiente. Es solo una vista (referencia) que apunta a los mismos datos que df. Esto genera dos problemas:

Si más adelante modificás algo en mufi (por ejemplo, cambiás un valor en una celda), corrés el riesgo de modificar también el df original sin querer, porque ambos "apuntan" a la misma memoria.
Pandas suele mostrar una advertencia llamada SettingWithCopyWarning, que avisa que no está claro si estás trabajando sobre una copia o sobre el original. Esto genera comportamiento impredecible: a veces modifica el original, a veces no.
