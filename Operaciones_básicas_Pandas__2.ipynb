## Indexación y slicing en pandas

Pandas pone a nuestra disposición nuevos métodos de indexación que permiten tener un mayor control sobre la misma y superar las limitaciones que nos impone NumPy sobre este tipo de estructuras.

Aclaración:

Indexar = Quiero el dato que está en esta dirección.La indexación es el mecanismo para acceder a un dato que ya sabés dónde está.

Filtrar = Quiero los datos que cumplan esta condición.El filtrado es cuando no sabés en qué fila está el dato, pero sabés qué característica tiene que cumplir

Slicing= Mientras que la indexación común suele ser para buscar un "punto" (un valor específico), el slicing se usa para obtener un segmento o tajada de la estructura.


Veamos todas las posibles combinaciones.<br/>

## Índices jerárquicos en pandas

Los índices jerárquicos de 'pandas' permiten tener más de un nivel en cualquiera de los índices de una estructura. Esto puede servir para agrupar más claramente los datos, o para conseguir identificar las filas por una clave única. En cierto modo, también es una forma de poder trabajar con tablas de más de dos dimensiones.

```python
peliculas = pd.DataFrame(
            {'Valoración':[6, None, 8.75, None],
             'Presupuesto':[160, 250, 100, None],
             'Director':['Peter Jackson', 'Gareth Edwards', 'Martin Scorsese', 'Alfonso Cuarón']},
            index = [[2014, 2014, 2013, 2013], ['Godzilla', 'El Hobbit III', 'El lobo de Wall Street', 'Gravity']]
)
peliculas
```

A partir de la construcción del índice jerárquico, podemos hacer indexaciones totales (mediante tuplas) o parciales (mediante selección de uno de los elementos del índice).

```python
# Indexación total
peliculas.loc[(2014, 'Godzilla')]
```

```python
# Indexación parcial
peliculas.loc[2014]
```

Podemos pasar niveles del índice jerárquico de las filas a las columnas mediante la función <b>unstack</b>, como si de una Pivot Table de Excel se tratase. El efecto: El DataFrame se hace más "bajo" y más ancho.

Con <b>stack</b> realizarmeos la operación contraria.

```python
peliculas
```

```python
# Pasamos el último nivel del índice de filas al de columnas
#peliculas_2 = peliculas.unstack()
# Rellenamos los NaNs con un guión o un cero para que no molesten visualmente
peliculas_2 = peliculas.unstack().fillna('-')
peliculas_2
```

```python
# La forma "moderna" de hacerlo
peliculas_2.stack(future_stack=True)
```

```python
peliculas
```

```python
# Pasa el nivel de las columnas al índice de las filas.
peliculas.stack()
```

## Modificación de índices en Pandas

```python
dataframe
```

```python
dataframe.reset_index()
```

Del mismo modo, podemos reestablecer un conjunto de columnas como índice de un DataFrame con la función <b>set_index</b>.

```python
dataframe
```

```python
dataframe.set_index(['c1'])
```
