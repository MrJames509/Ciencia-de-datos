##Desafío de Datos: Pandas a Prueba

El Dataset de trabajo:
Copiá y ejecutá este código en tu celda de Python para empezar:

```python
import pandas as pd
import numpy as np

data = np.arange(16).reshape(4, 4)
df = pd.DataFrame(data, index=['f1', 'f2', 'f3', 'f4'], columns=['c1', 'c2', 'c3', 'c4'])
# El DataFrame resultante tiene valores del 0 al 15
```

Tu Misión:

Escribí el código necesario para extraer exactamente los siguientes resultados del objeto df:

Resultado A: El valor único 5.

Resultado B: Una Serie que contenga solo los valores 0, 1, 2, 3.

Resultado C: Un DataFrame que solo tenga las columnas c1 y c3 (con todas sus filas).

Resultado D: Los valores del 4 al 11 (filas f2 y f3 completas).

Resultado E: Solo los valores que sean pares (múltiplos de 2) dentro de todo el DataFrame.

Resultado F: Un recorte que muestre el cuadrado central: valores 5, 6, 9, 10.

Resultado G: Los valores de las últimas dos filas (f3 y f4) pero solo de las columnas c1 y c4.

Para la entrega:


Para cada resultado, debés escribir el código de dos formas distintas: una usando etiquetas (.loc) y otra usando posiciones numéricas (.iloc)

```python
# Resultado A: El valor único 5
res_A_loc = df.loc['f2', 'c2']
res_A_iloc = df.iloc[1, 1]

print("A (loc):", res_A_loc)
print("A (iloc):", res_A_iloc)
```

```python
# Resultado B: Una Serie que contenga solo los valores 0, 1, 2, 3
res_B_loc = df.loc['f1']
res_B_iloc = df.iloc[0]

print("B (loc):\n", res_B_loc, sep="")
print("\nB (iloc):\n", res_B_iloc, sep="")
```

```python
# Resultado C: Un DataFrame con las columnas c1 y c3 (todas sus filas)
res_C_loc = df.loc[:, ['c1', 'c3']]
res_C_iloc = df.iloc[:, [0, 2]]

print("C (loc):\n", res_C_loc, sep="")
print("\nC (iloc):\n", res_C_iloc, sep="")
```

```python
# Resultado D: Los valores del 4 al 11 (filas f2 y f3 completas)
res_D_loc = df.loc['f2':'f3']
res_D_iloc = df.iloc[1:3]

print("D (loc):\n", res_D_loc, sep="")
print("\nD (iloc):\n", res_D_iloc, sep="")
```

```python
# Resultado E: Solo los valores que sean pares dentro de todo el DataFrame

# Con .loc: Usamos .where() para filtrar manteniendo la estructura de etiquetas
res_E_loc = df.where(df % 2 == 0)

res_E_iloc = df.iloc[lambda x: (x % 2 == 0).values]

print("E (loc):\n", res_E_loc, sep="")
print("\nE (iloc):\n", res_E_iloc, sep="")
```

```python
# Resultado F: Un recorte que muestre el cuadrado central (valores 5, 6, 9, 10)
res_F_loc = df.loc['f2':'f3', 'c2':'c3']
res_F_iloc = df.iloc[1:3, 1:3]

print("F (loc):\n", res_F_loc, sep="")
print("\nF (iloc):\n", res_F_iloc, sep="")
```

```python
# Resultado G: Las últimas dos filas (f3 y f4) para las columnas c1 y c4
res_G_loc = df.loc['f3':'f4', ['c1', 'c4']]
res_G_iloc = df.iloc[2:4, [0, 3]]

print("G (loc):\n", res_G_loc, sep="")
print("\nG (iloc):\n", res_G_iloc, sep="")
```
